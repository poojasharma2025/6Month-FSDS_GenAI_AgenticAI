```python
import pandas as pd
import seaborn as sns
import ollama
import matplotlib.pyplot as plt

# load the titanic dataset
url=r"C:\Users\WELCOME\Documents\Data Science\July 2025\28th july-titanic EDA_LLM\titanic_ dataset_final.csv"
df=pd.read_csv(url)
df
```

```python
def generate_insights(df_summary):
    prompt=f"analyze the dataset summary and provide insight:\n\n{df_summary}"
    response=ollama.chat(model="mistral",messages=[{"role":"user","content":prompt}])
    return response['message']['content']

#Generate AI insight
summary = df.describe().to_string()
insights = generate_insights(summary)
print("\n ▪ AI-GENERATED INSIGHT:\n",insights)
```

```python
import gradio as gr
def eda_analysis(file):
    df=pd.read_csv(file.name)
    summary = df.describe().to_string()
    insights = generate_insights(summary)
    return insights
# create web interface
demo = gr.Interface(fn=eda_analysis,inputs="file",outputs="text",title="AI powered EDA with mistral")

# launch App
demo.launch(share=True)  # use share true for google colab
```
